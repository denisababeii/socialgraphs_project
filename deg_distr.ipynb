Degree distributions

```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

network = nx.DiGraph()

# Load nodes
nodes_df = pd.read_csv('/Volumes/MASTER/Year1/Social_graphs/project/Nodes.csv')
for index, row in nodes_df.iterrows():
    network.add_node(row['Id'], **row.to_dict())

# Load edges
edges_df = pd.read_csv('/Volumes/MASTER/Year1/Social_graphs/project/Edges.csv', delimiter=';')
for index, row in edges_df.iterrows():
    network.add_edge(row['Source'], row['Target'])
```

```python
print(network.number_of_nodes())
print(network.number_of_edges())

print(max(dict(network.in_degree()).values()))
print(min(dict(network.in_degree()).values()))

print(max(dict(network.out_degree()).values()))
print(min(dict(network.out_degree()).values()))

counts, bin_edges = np.histogram([deg for node, deg in network.in_degree()], 500)
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])

plt.figure()
plt.plot(bin_centers, counts, 'o')
plt.xlabel('In Degree')
plt.ylabel('Count')
plt.title('In Degree Distribution (Linear)')
plt.show()

plt.figure()
plt.plot(bin_centers, counts, 'o')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('In Degree (log)')
plt.ylabel('Count (log)')
plt.title('In Degree Distribution (Log-Log)')
plt.show()

counts, bin_edges = np.histogram([deg for node, deg in network.out_degree()], 500)
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])

plt.figure()
plt.plot(bin_centers, counts, 'o')
plt.xlabel('Out Degree')
plt.ylabel('Count')
plt.title('Out Degree Distribution (Linear)')
plt.show()

plt.figure()
plt.plot(bin_centers, counts, 'o')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Out Degree (log)')
plt.ylabel('Count (log)')
plt.title('Out Degree Distribution (Log-Log)')
plt.show()
```
